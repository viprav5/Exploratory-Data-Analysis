## Exploratory Analysis and Visualization

After data preparation and cleaning we will plot graphs for better visualization and understanding, this can be done by importing matplotlib and seaborn libraries of python. These libraries have many functions for ploting graphs between datas of different columns as necessary, which we wil discuss below.



Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### Creating dataframe of total recovery per month.

```python
re_df = covid_df.groupby('Month')[['Recovered']].sum() 
re_df
```

###  Creating dataframe of total confirmed cases per month

```python
con_df = covid_df.groupby('Month')[['Confirmed']].sum() 
con_df
```

### Merge the above dataframes to get record of recovered and confimed cases in single dataframe

```python
rc_df=pd.merge(re_df,con_df,left_index=True,right_index=True) 
rc_df
```

###  Storing recovery percent in new column

```python
rc_df['RecoverConfirmPer']=(rc_df['Recovered']/rc_df['Confirmed'])*100   
rc_df
```

#### 1. barplot() of seaborn is used to plot bar graph according to the data provided. Here the plot shows the           recoverd and deaths per month ,it can be computed by bar heights.

```python
sns.barplot(x=covid_df.Month,y=covid_df.Recovered,color='b',saturation=0.3,label='Rcovered')
sns.barplot(x=covid_df.Month,y=covid_df.Deaths,color='orange',saturation=0.3,label='Deaths')
plt.ylabel('No. of people.')
plt.title("Month VS Recovered & Death")
plt.legend();
```

#### 2. Now we will plot the line graph using matplotlib functions,to analyse the recovery percent according to months.

```python
plt.xlabel('Month')
plt.ylabel('Percent')
plt.title('Recovery Percentage')
plt.plot(rc_df.index,rc_df.RecoverConfirmPer,'x-r');
```

#### 3. scatterplot()  is a function of seaborn lib used to plot with possibility of several semantic groupings. Here scatterplot is ploted according to months.

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x=covid_df.Recovered,y=covid_df.Deaths,data=covid_df,hue='Month',palette='viridis',style='Month',s=100)
plt.ticklabel_format(style = 'plain');
```

#### 4. Now we will extract total recovered and total active cases and uses plot() function of matplotlib to draw line graph between recovered and active cases per month.

```python
recover_df = covid_df.groupby('Month')[['Recovered']].sum()
recover_df
```

```python
active_df = covid_df.groupby('Month')[['Active']].sum()
active_df
```

```python
plt.plot(active_df.index,active_df.Active,'x--b',label='Active case');
plt.plot(active_df.index,recover_df.Recovered,'o--r',label='Recovered');
plt.ticklabel_format(style = 'plain')
plt.xlabel('Months')
plt.title("Active and Recovered cases per Month")
plt.legend();
```

#### Finding total deaths per month

```python
death_df = covid_df.groupby('Month')[['Deaths']].sum()
death_df
```

#### 5. distplot function of seaborn is a combination of hist() and rungplot(). It gives univariate distribution of total deaths/month observations.

```python
sns.distplot(death_df.Deaths,rug=True,kde=True,rug_kws={'color':'r','height':0.2},bins=5)
plt.ticklabel_format(style = 'plain');
```

## Asking and Answering Questions

In this section we are using some intresting functions to explore the dataset such as grouby,merge,sort_values,drop_duplicates and accessing data using loc[].



### Q1: Find top 10 countries which have highest deaths and plot bar graph showing number of deaths in those countries ?

```python
country_death_df = covid_df.groupby("Country")[['Deaths']].sum()
country_death_df = country_death_df.sort_values('Deaths',ascending=False).head(10)
country_death_df
```

```python
plt.figure(figsize=(18,7))
sns.barplot(x=country_death_df.index,y=country_death_df.Deaths)
plt.title('Top 10 countries with highest no. of deaths',fontsize=30)
plt.xlabel('Countries',fontsize=25)
plt.ylabel('Total deaths',fontsize=25)
plt.xticks(rotation=45)
plt.ticklabel_format(style = 'plain',axis='y')
plt.show()
```

#### Q2: Find the maximum number of recovery and the month also? 

```python
recover_df = covid_df.groupby('Month')[['Recovered']].sum()
```

```python
recover_df.sort_values('Recovered',ascending=False).head(1)
```

Here we have used groupby() which creats the group according to arg passed in it ,here it group data by month and another sort_values() is used to sort the values in desccending order as (ascending=False).

#### Q3: Find minimum number of deaths in a month?

```python
death_df = covid_df.groupby('Month')[['Deaths']].sum()
```

Here we have used groupby() which creats the group according to arg passed in it ,here it group data by deaths and another sort_values() is used to sort the values in ascending order and head(1) returns the first record.

#### Q4: Access the records having confirmed cases less than 100 and recovery is grater than 10?

```python
new_df = covid_df.drop_duplicates(subset=['Confirmed'])
new_df
```

```python
new_df.loc[(new_df.Confirmed<100) & (new_df.Recovered>10)]
```

Query is passed in loc according to query it returns the data frame.

#### Q5: Built a dataframe having number of deaths less than 10 and recovered cases less than equal to 100?

```python
df=new_df.loc[(new_df.Deaths<10) & (new_df.Recovered<=100)]
df
```

In this query, it holds two conditions using (&) operator and it will return those data which will satisfy both the conditions.
