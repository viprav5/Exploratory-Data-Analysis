# Google Play Store Data Analysis.

This dataset contains the data of applications available on google play store and it was created in 2018-2019. This dataset contains
10841 rows and 13 columns.
 

## Columns Description

* App - Application name, there are 9660 unique apps.

* Category - Category the app belongs to and there are 34 categories.

* Rating - Overall user rating of the app (as when scraped).

* Reviews - Number of user reviews for the app (as when scraped).

* Size - Size of the app (as when scraped).

* Installs - Number of user downloads/installs for the app (as when scraped).

* Type - There are two types Paid or Free.

* Price - Price of the app (as when scraped).

* Content Rating - Age group the app is targeted at - Children / Mature 21+ / Adult.

* Genres - An app can belong to multiple genres (apart from its main category).

* Last Updated - Date when the app was last updated on Play Store (as when scraped).

* Current Ver - Current version of the app available on Play Store (as when scraped).

* Android Ver - Android Ver - Min required Android version (as when scraped).

## Data Preparation and Cleaning

To read dataset we have to import pandas and by using read_csv() function of pandas we can read data which have csv format. After reading we determined it's shape i.e rows and columns ,it's information. We can also find mean ,median ,mode etc. After that we used drop() function to remove coloumns which are not needed for analysis.

```python
import pandas as pd
```

### Loading dataset

```python
playstore_df=pd.read_csv(data_dir + "/googleplaystore.csv")
playstore_df
```

### Using drop() function we remove columns not requred for analysis.

```python
playstore_df = playstore_df.drop(['Last Updated','Current Ver','Genres'],axis=1)
```

```python
playstore_df
```

```python
playstore_df.Category.unique()
```

### Dropping '1.9' from category column as there is no '1.9' category on Play Store.

```python
playstore_df[playstore_df.Category=='1.9']
```

```python
playstore_df=playstore_df.drop(index=10472)
```

```python
playstore_df.Type.unique()
```

### Determing Null values in each columns

```python
playstore_df.info()
```

```python
playstore_df.mean()
```

### Filling null values in rating column by round-off of mean of rating i.e. '4.2'

```python
playstore_df['Rating'] = playstore_df.Rating.fillna(4.2)  
```

```python
playstore_df.Rating.unique()
```

```python
playstore_df.head()
```

```python
playstore_df=playstore_df.dropna()
```

```python
playstore_df.describe()
```

### Converting Reviews columns to numeric

```python
playstore_df.Reviews = pd.to_numeric(playstore_df.Reviews)
```

```python
playstore_df.Price.unique()
```

### Converting Price into numeric we have to remove '$' symbol from price

```python
playstore_df['Price'] = pd.to_numeric(playstore_df['Price'].str.replace('$',''))
```

```python
playstore_df.reset_index(drop=True)
```

### Removing symbols ('+' , ',') from Installs column and converting it into numeric

```python
playstore_df['Installs'] = playstore_df['Installs'].str.replace('+','')
playstore_df.head()
```

```python
playstore_df['Installs'] = pd.to_numeric(playstore_df['Installs'].str.replace(',',''))
playstore_df.head()
```

```python
playstore_df = playstore_df.sort_values('Reviews',ascending=False)
```

### Droping duplicate App entries in dataset

```python
playstore_df = playstore_df.drop_duplicates('App').reset_index(drop=True)
playstore_df.head()
```

```python
playstore_df.info()
```

## Exploratory Analysis and Visualization

After data preparation and cleaning we will plot graphs for better visualization and understanding, this can be done by importing matplotlib and seaborn libraries of python. These libraries have many functions for ploting graphs between datas of different columns as necessary, which we wil discuss below.

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### Deriving a dataframe containing total no. of installs per category and sorting them in ascending order

```python
total_category_df = playstore_df.groupby('Category')[['Installs']].sum()
total_category_df = total_category_df.sort_values('Installs',ascending=False).head(20)
total_category_df
```

### 1. barplot() of seaborn is used to plot bar graph according to the data provided. Here the plot shows the total number of apps of per category installs, it can be computed by bar heights.

```python
plt.figure(figsize=(14,7))
plt.title('Category VS Total Installs',fontsize=30)
sns.barplot(x=total_category_df.index,y=total_category_df.Installs)
plt.xticks(rotation=70)
plt.xlabel('Category',fontsize=30)
plt.ylabel('Total Installs',fontsize=30)
plt.show()
```

### 2. Using pie() we are ploting the piechart and depicting the total percent of apps coming under different content rating

```python
playstore_df['Content Rating'].value_counts(sort=False).plot.pie(autopct='%1.2f%%',explode=[0.2,0.2,0,0,0.3,0],figsize=(30,15))
plt.legend();
```

```python
plt.title('Content Rating vs Numbers of apps')
plt.xlabel("Content Rating")
plt.ylabel("Numbers of apps")
playstore_df['Content Rating'].value_counts().plot(kind='bar');
```

### 3. barplot() of seaborn is used to plot bar graph according to the data provided. Here the plot shows the ratings of free and paid apps per category

```python
type_df = playstore_df.groupby(['Type','Category'])[['Rating']].mean()
type_df
```

```python
free_df = type_df.loc['Free']
paid_df = type_df.loc['Paid']
merged_df = pd.merge(free_df,paid_df,on='Category',suffixes=('_free','_paid'))
merged_df.head()
```

```python
merged_df['Rating Diff'] = merged_df['Rating_free']-merged_df['Rating_paid']
merged_df.head()
```

```python
plt.figure(figsize=(14,4))
plt.xticks(rotation=90)
plt.title('Category vs Free_Apps_rating')
sns.barplot(x=merged_df.index,y=merged_df.Rating_free)
plt.show()
plt.figure(figsize=(14,4))
plt.xticks(rotation=90)
plt.title('Category vs Paid_Apps_rating')
sns.barplot(x=merged_df.index,y=merged_df.Rating_paid)
plt.show()
plt.figure(figsize=(14,7))
plt.xticks(rotation=90)
plt.title('Category vs Difference between Paid And Free Apps',fontsize=20)
sns.barplot(x=merged_df.index,y=merged_df['Rating Diff']);
```

### 4. plot() of matplotlib is used to plot the line graph, here we are plotting Content Rating VS Total Installs graph

```python
cnt_ins_df = playstore_df.groupby('Content Rating')[['Installs']].sum()
cnt_ins_df
```

```python
plt.figure(figsize=(14,5))
plt.xlabel('Content Rating')
plt.ylabel('Total Installs')
plt.title('Content Rating VS Total Installs')
plt.plot(cnt_ins_df.index,cnt_ins_df.Installs,'x--r')
plt.ticklabel_format(style='plain',axis='y');
```

### 5. pairplot() : To plot multiple pairwise bivariate distributions in a dataset

```python
sns.pairplot(playstore_df, hue='Content Rating', height=2.5);
```

#### Q1: Considering reviews and ratings, which factor has more impact on the number of installations?

```python
rr_df = playstore_df.groupby('Category')[['Installs','Reviews','Rating']].mean()
rr_df.head()
```

```python
q = rr_df.corr()
sns.heatmap(q,annot=True);
```

## From heatmap we can analyze that from reviews and ratings, reviews have more impact on the number of installations.
#### -So positive and more reviews will increase installations

### Q2: Find top 10 apps with highest installs ?
